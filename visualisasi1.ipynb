```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 5)
```

```python
father_son = pd.read_csv('datasetmap/father_son.csv', usecols=['fheight', 'sheight'])
father_son.head()
```

```python
plt.scatter(father_son.fheight, father_son.sheight, c='darkred', marker='s');
plt.grid(color='black');
```

```python
plt.scatter(father_son.fheight, father_son.sheight, c = 'yellow', edgecolor= 'black');
plt.grid(color='black');
```

```python
plt.scatter(father_son.fheight, father_son.sheight,  c = 'yellow', edgecolor = 'black');
plt.grid(color='black');
plt.xlabel('father height (inches)');
plt.ylabel('son height (inches)');
plt.title('Son Height as a Function of Father Height');
```

```python
import geopandas as gpd
service_district = gpd.read_file('datasetmap/ServiceDistricts.shp')
print(service_district.head())

print(service_district.loc[0, 'geometry'])
```

```python
service_district.to_crs(epsg=3857).plot(figsize=(10, 10));
service_district.to_crs(epsg=3857).plot(column='Name', legend=True, figsize=(10, 10));
```

```python
service_district.to_crs(epsg=4326).plot(column='Name', figsize=(10, 10));
plt.scatter(x=chickens.lng, y=chickens.lat, c='black');
plt.grid();
```

```python
service_district.to_crs(epsg=4326).plot(column='Name', legend=True);
plt.scatter(x=chickens.lng, y=chickens.lat, c='yellow', edgecolor='black');

plt.title('Nashville Chicken Permits');
plt.xlabel('longitude');
plt.ylabel('latitude');
plt.grid();
```
